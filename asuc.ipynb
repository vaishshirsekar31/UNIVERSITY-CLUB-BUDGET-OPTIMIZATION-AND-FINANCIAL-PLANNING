# ASUC Budget by Organization Type

```python
asuc_groups = [
    'ASUC CHARTERED PROGRAMS AND COMMISSIONS',
    'PUBLICATION GROUPS',
    'STUDENT ACTIVITY GROUPS',
    'STUDENT-INITIATED SERVICE GROUPS',
    'GRANTS, SCHOLARSHIPS, & FUNDS',
    'OFFICE OF THE PRESIDENT',
    'OFFICE OF THE EXECUTIVE VICE PRESIDENT',
    'OFFICE OF THE EXTERNAL AFFAIRS VICE PRESIDENT',
    'OFFICE OF THE STUDENT ADVOCATE',
    'SENATE',
    'APPOINTED OFFICIALS',
    'OPERATIONS',
    'ELECTIONS',
    'EXTERNAL EXPENDITURES'
]

# pandas processing to get each group
org_type = asuc_budget.groupby(['Sponsorship Category Name'])

# array of DataFrames for each group
asuc_group_dataframes = []
for g in asuc_groups:
    asuc_group_dataframes.append(org_type.get_group(g))

for i in range(len(asuc_groups)):
    print(asuc_groups[i])
    plt.figure()
    asuc_group_dataframes[i].plot.bar(x=name, y=allocation)
    plt.show()
```

```python
print('Sum of $ / Group')

y_amounts = []
for i in range(len(asuc_groups)):
    y_amounts.append(asuc_group_dataframes[i]['Initial Allocation ($)'].sum())
    
print(y_amounts)
plt.figure()
plt.bar(np.arange(len(asuc_groups)), y_amounts)
plt.show()
```

# Money Based on Age of Club (# of Years)

```python
clubs_with_years = asuc_budget.loc[asuc_budget['Year'].notnull()]

years_dataframe = clubs_with_years[['Year', 'Initial Allocation ($)']]
years_dataframe = years_dataframe.sort_values('Year')

years_dataframe.plot.scatter(x='Year', y='Initial Allocation ($)')
plt.show()
```
